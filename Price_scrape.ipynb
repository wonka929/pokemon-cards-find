```python
from bs4 import BeautifulSoup as bs
import requests
import warnings
import os

warnings.simplefilter(action="ignore", category=FutureWarning)
import pandas as pd

import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def url_to_image(url):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    # resp = urllib.request.urlopen(url)
    # resp = requests.get(url)
    # img = io.imread(url)
    resp = requests.get(url, stream=True).raw
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    # image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    # return the image
    return image
```

```python

for file in os.listdir("./"):
    try:
        if file.endswith("png"):
            
            r = requests.get("https://pokemoncardvalue.com/pokemon/" + file.split("_")[0] + "-pokemon-card/")
            contenuto = bs(r.text)
            
            elenco = pd.DataFrame()
            
            for elem in contenuto.findAll("div", {"class": "row panel"}):
                try:
                    names = elem.findAll("span", {"id": "no-q"})
                    # print(names[0].text)
                    images = elem.findAll("img", {"class": "card-image-resize"})
                    # print((images[0])["src"])
                    values = elem.findAll("table")
                    value = values[0].findAll("td")
                    # print(value[0].text + ", " + value[1].text + ", " + value[2].text)
            
                    dict = {
                        "Nome": names[0].text,
                        "Immagine": (images[0])["src"],
                        "Minimo": float(value[0].text[1:]),
                        "Media": float(value[1].text[1:]),
                        "Massimo": float(value[2].text[1:]),
                    }
            
                    elenco = elenco.append(dict, ignore_index=True)
                except:
                    continue
                    
            #elenco
            
            
            for url in elenco.Immagine.values:
            
                try:
                    #print(url)
                    img1 = url_to_image(url)
            
                    # req = urlopen(url)
                    # img1 = np.asarray(bytearray(req.read()), dtype=np.uint8)
                    img2 = cv2.imread(file, 4)
            
                    # Initiate SURF detector
                    surf = cv2.xfeatures2d.SURF_create()
            
                    # find the keypoints and descriptors with SURF
                    kp1, des1 = surf.detectAndCompute(img1, None)
                    kp2, des2 = surf.detectAndCompute(img2, None)
            
                    # BFMatcher with default params
                    bf = cv2.BFMatcher()
                    matches = bf.knnMatch(des1, des2, k=2)
            
                    how_many_found = 0
            
                    # Apply ratio test
                    good = []
                    for m, n in matches:
                        if m.distance < 0.75 * n.distance:
                            good.append([m])
                            a = len(good)
                            percent = (a * 100) / len(kp2)
                            # print("{} % similarity".format(percent))
                            if percent * 100 >= 75.00:
                                print("Match Found")
                                img3 = cv2.drawMatchesKnn(img1, kp1, img2, kp2, good, None, flags=2)
                                img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
                                plt.imshow(img3), plt.show()
                                print(elenco[elenco["Immagine"] == url])
                                ++how_many_found
                                break
            
                    # img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=2)
                    # plt.imshow(img3),plt.show()
            
                except Exception as e:
                    print(e)
                    continue
            
            
            if how_many_found == 0:
                print("Non è stato prossibile trovare un match")
    except:
        continue
```
